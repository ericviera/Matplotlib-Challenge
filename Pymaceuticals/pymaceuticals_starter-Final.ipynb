## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np 

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
comb_df = pd.merge(mouse_metadata,study_results, on='Mouse ID')
# Display the data table for preview
comb_df
```

```python
# Checking the number of mice.
comb_df["Mouse ID"].count()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

comb_dp=comb_df[comb_df.duplicated(['Mouse ID','Timepoint'])]
comb_dp
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
comb_drp=comb_df.drop_duplicates(['Mouse ID','Timepoint'])
comb_drp
```

```python
# Checking the number of mice in the clean DataFrame.
comb_drp['Mouse ID'].count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summ=comb_drp['Drug Regimen'].unique()
summ
# This method is the most straighforward, creating multiple series and putting them all together at the end.
#comb_drp['Tumor Volume (mm3)'].describe()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
summ=pd.DataFrame(comb_drp.groupby('Drug Regimen').count())
summ
```

```python
summ=pd.DataFrame(comb_drp.groupby('Drug Regimen').count())
summ["Mean"] = pd.DataFrame(comb_drp.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean())
summ["Median"] = pd.DataFrame(comb_drp.groupby("Drug Regimen")["Tumor Volume (mm3)"].median())
summ["Variance"] = pd.DataFrame(comb_drp.groupby("Drug Regimen")["Tumor Volume (mm3)"].var())
summ["Standard Deviation"] = pd.DataFrame(comb_drp.groupby("Drug Regimen")["Tumor Volume (mm3)"].std())
summ["SEM"] = pd.DataFrame(comb_drp.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem())
summ = summ[['Mouse ID','Mean','Median','Variance','Standard Deviation','SEM']]
summ
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 

total_mice= summ[['Mouse ID']]
total_mice.plot(kind="bar")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
Capomulin_treatment = comb_df.loc[(comb_df["Drug Regimen"] == "Capomulin"),:]
Capomulin_treatment
```

```python
timepoint = Capomulin_treatment["Timepoint"]
tumor_vol = Capomulin_treatment["Tumor Volume (mm3)"]
tumor_volume_line, = plt.plot(timepoint, tumor_vol)
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')

plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
avg_weight = Capomulin_treatment.groupby(Capomulin_treatment["Mouse ID"])["Weight (g)"].mean()
avg_tumor_vol = Capomulin_treatment.groupby(Capomulin_treatment['Mouse ID'])["Tumor Volume (mm3)"].mean()
plt.scatter(avg_weight,avg_tumor_vol)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
avg_weight = Capomulin_treatment.groupby(Capomulin_treatment["Mouse ID"])["Weight (g)"].mean()
avg_tumor_vol = Capomulin_treatment.groupby(Capomulin_treatment["Mouse ID"])["Tumor Volume (mm3)"].mean()


slope, intercept, r, p, std_err = st.linregress(avg_weight, avg_tumor_vol)
regress_values = avg_weight * slope + intercept
plt.scatter(avg_weight,avg_tumor_vol)
plt.xlabel("Average Weight")
plt.ylabel("Average Volume")
plt.plot(avg_weight,regress_values,"--")
plt.xticks(avg_weight, rotation=90)
plt.show()
```
